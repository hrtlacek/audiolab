```python
N = 50
imp = zeros(N)
imp[10] = 1
grid(True)
_ = stem(imp)
```

```python
x = imp
delay = 2
delLen = 10
delBuf = zeros(delLen)
y = zeros_like(x)
for n in range(len(x)):
    delayed = delBuf[(n-delay)%delLen]
    delBuf[n%delLen] = x[n] +delayed*0.8
    y[n] = delayed
    
```

```python
stem(y)
grid(True)
```
