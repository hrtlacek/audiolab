## Linear to DB Conversion

````python
def linToDb(linearValues):
    """
    Convcert Linear Values to dB
    """
    dbValues =20*np.log10(linearValues)
    return dbValues
````

````python
linToDb(0.5)
````

## Plotten von input-output unserer neuen Funktion

````python
inputValues = np.linspace(0.0001, 1, 100) #erzeugt 100 float zahlen zwischen 0.0001 und 1
plt.plot(inputValues,linToDb(inputValues))
plt.title('Linear to db Test')
plt.grid()
plt.xlabel('linear')
plt.ylabel('[dB]')
plt.show()
````

## For schleifen in python
For schleifen in python sollten eigentlich `foreach` schleifen heissen. 
```python

    ...
``` 
iteriert über die Liste, in jedem durchlauf erhält `eineVariable` den Wert des nächsten Elements in der Liste.

````python
meineListe = ['hans', 'peter', 'klaus']
for etwas in meineListe:
    print(etwas)
````

## Beispiel: File zerschnipseln

````python
x,sr = librosa.load('TestAudio.wav', sr=None) #Nur um hier alles an einem Ort zu haben: wir laden nocheinmal das audio file.
anzahlSchnipsel = 5
N = int(np.shape(x)[0]) #Num Samples
laengeSchnipsel = int(N/anzahlSchnipsel) #floor operation (comma wegschmeissen)
for i in range(anzahlSchnipsel):
    print('wert von i', i)
    schnipsel = x[laengeSchnipsel*i:laengeSchnipsel*(i+1)]
    myOutPath = 'schnipsel_'+str(i)+'.wav'
    librosa.output.write_wav(myOutPath,schnipsel, sr,norm=True)
    plt.plot(schnipsel)
    plt.show()
    plt.specgram(schnipsel)
    plt.show()
````


### Unterschiedlich Lange arrays übereinander plotten
Es ist die idee der Hü, das audio file und den laufenden RMS übereinander zu plotten. Die Liste/der array der RMS werte wird am ende eine andere(kürzere) Länge aufweisen als euer audio file.

um die zeitpunkte der einzelnen werte anzugleichen müssen zwei Zeitvariablen gemacht werden die als X-Achsen dienen. 

````python
Nlang = 30 #beispiel fuer anzahl der audio samples
Nkurz = 10 #beispiel fuer anzahl der RMS werte
T = 10 # Zeit in sekunden (muesste durch laenge in samples/sample rate berechnet werden)
tSamps = np.linspace(0,T,Nlang) #generierung von zeit array mit gleicher sample rate wie audio (gleich vielen werten)
tRms = np.linspace(0,T,Nkurz) #generierung von zeit array mit gleicher sample rate wie rms werterfassung (gleich vielen werten wie rms array in der hue)

kurzerArray = np.random.random(Nlang) #generierung von beispiel signal (audio)
langerArray = np.random.random(Nkurz) #generierung von beispiel signal (rms)
plt.plot(tSamps,kurzerArray) #plotten von langem zeit-array gegen langes Signal
plt.plot(tRms,langerArray) #plotten von kurzem zeit-array gegen kurzes Signal
plt.show()
````

### Wieso so kompliziert?
Hier was passiert wenn wir uns nicht darum kümmern: also wie es **nicht** geht:

````python
plt.plot(kurzerArray) #plotten von langem zeit-array gegen langes Signal
plt.plot(langerArray) #plotten von kurzem zeit-array gegen kurzes Signal
plt.show()
````
