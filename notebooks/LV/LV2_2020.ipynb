# Cubic Distortion

https://ccrma.stanford.edu/realsimple/faust_strings/Cubic_Nonlinear_Distortion.html

$
f(x) = \left\{\begin{array}{ll}
-\frac{2}{3}, & x\leq -1 \\ 
x-\frac{x^3}{3}, & -1 < x < 1 \\ 
\frac{2}{3}, & x\geq 1. \\
\end{array} \right.
$


```python
def myDist(x):
    x = x +0.3
    c = clip(x,-1,1)
    y = c-(c**3)/3
    return y
```

```python
x = linspace(-2, 2, 100)
```

```python
y = myDist(x)
plot(x, y)
grid()
ylim([-0.8, 0.8])
```

```python
def meinPlotSpec(x, sr= 44100):
    N = len(x)
    faxis = linspace(0, sr/2, N//2)
    spectrum = abs(fft.fft(x))[0:N//2]
    plot(faxis, spectrum)
    xlim([500, 10000])
```

```python
sr =44100
T = 5 #sekunden
N = 5*sr # anzahl d samples
n = arange(N) # sample index
t = n/sr # zeit index in sekunden

mySin = sin(t *1000 * 2 *pi)
```

```python
meinOutp = myDist(mySin*10)
meinPlotSpec(meinOutp, sr= sr)
```

```python
meinPlotSpec(myDist(mySin*10))
```

```python
subplot(121)
meinPlotSpec(myDist(mySin))

subplot(122)
meinPlotSpec(myDist(mySin*50))
```
